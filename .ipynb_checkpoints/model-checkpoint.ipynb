```python
#Now we will start modelling
```

```python
x=df.drop(columns="target")
y=df.target.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Linear support vector classifier
from sklearn.svm import LinearSVC
LinearSVC=LinearSVC()
```

```python
LinearSVC.fit(x_train,y_train)
yp_LinearSVC=LinearSVC.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,f1_score,recall_score
confusion_matrix(y_test,yp_LinearSVC)
```

```python
AccuracySVC=accuracy_score(y_test,yp_LinearSVC)
print("Accuracy =",accuracy_score(y_test,yp_LinearSVC))
print("Precision =",precision_score(y_test,yp_LinearSVC))
print("Recall =",recall_score(y_test,yp_LinearSVC))
print("F1 =",f1_score(y_test,yp_LinearSVC))
```

```python
#K Nearest Neighbour Classifier
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
yp_KNN=KNN.predict(x_test)
print("Confusion Matrix =",confusion_matrix(y_test,yp_KNN))
AccuracyKNN=accuracy_score(y_test,yp_KNN)
print("Accuracy =",accuracy_score(y_test,yp_KNN))
print("Precision =",precision_score(y_test,yp_KNN))
print("Recall =",recall_score(y_test,yp_KNN))
print("F1 =",f1_score(y_test,yp_KNN))
```

```python
#Random Forest CLassifier
from sklearn.ensemble import RandomForestClassifier
randomForest=RandomForestClassifier()
randomForest.fit(x_train,y_train)
yp_RandomForest=randomForest.predict(x_test)
print("Confusion Matrix =",confusion_matrix(y_test,yp_RandomForest))
AccuracyRandomForest=accuracy_score(y_test,yp_RandomForest)
print("Accuracy =",accuracy_score(y_test,yp_RandomForest))
print("Precision =",precision_score(y_test,yp_RandomForest))
print("Recall =",recall_score(y_test,yp_RandomForest))
print("F1 =",f1_score(y_test,yp_RandomForest))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
LogisticRegression=LogisticRegression()
LogisticRegression.fit(x_train,y_train)
yp_LogisticRegression=LogisticRegression.predict(x_test)
print("Confusion Matrix =",confusion_matrix(y_test,yp_LogisticRegression))
AccuracyLogisticRegression=accuracy_score(y_test,yp_LogisticRegression)
print("Accuracy =",accuracy_score(y_test,yp_LogisticRegression))
print("Precision =",precision_score(y_test,yp_LogisticRegression))
print("Recall =",recall_score(y_test,yp_LogisticRegression))
print("F1 =",f1_score(y_test,yp_LogisticRegression))
```

```python
acc=[AccuracySVC,AccuracyKNN,AccuracyRandomForest,AccuracyLogisticRegression]
models=["SVC","KNN","Random Forest", "Logistic Regression"]
plt.figure(figsize=(10,5))
plt.bar(models,acc)
plt.title("Different Models and there accuracies")
plt.xlabel("Model")
plt.ylabel("Accuracy")
```

```python
#Now we start hypertuning
```

```python
train_score=[]
test_score=[]
KNNModel2=KNeighborsClassifier()
```

```python
for i in range(1,21):
    KNNModel2.set_params(n_neighbors=i)
    KNNModel2.fit(x_train,y_train)
    train_score.append(KNNModel2.score(x_train,y_train))
    test_score.append(KNNModel2.score(x_test,y_test))
```

```python
neighbors=range(1,21)
plt.plot(neighbors,train_score,label="Training Accuracy")
plt.plot(neighbors,test_score,label="Test Accuracy")
plt.xticks(np.arange(1,21,1))
plt.xlabel("Neighbours(K)")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Training and Test accuracy for different values of K")
```

```python
print("Train accuracies",train_score,'\n')
print("Test accuracies",test_score,'\n')
```

```python
#K=11 or n_neighbors=11 seems best
```

#RandomisedSearchCV

```python
# Different LogisticRegression hyperparameters
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

# Different RandomForestClassifier hyperparameters
rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Setup random seed
np.random.seed(42)

# Setup random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression,
                                param_distributions=log_reg_grid,
                                cv=5,
                                n_iter=20,
                                verbose=True)

# Fit random hyperparameter search model
rs_log_reg.fit(x_train, y_train);
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test,y_test)
```

```python
# Setup random seed
np.random.seed(42)

# Setup random hyperparameter search for RandomForestClassifier
rs_rf = RandomizedSearchCV(randomForest,
                           param_distributions=rf_grid,
                           cv=5,
                           n_iter=20,
                           verbose=True)

# Fit random hyperparameter search model
rs_rf.fit(x_train, y_train);
```

```python
# Find the best parameters
rs_rf.best_params_
```

```python
# Evaluate the randomized search random forest model
rs_rf.score(x_test, y_test)
```

Tuning model with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Different LogisticRegression hyperparameters
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

# Setup grid hyperparameter search for LogisticRegression
gs_log_reg = GridSearchCV(LogisticRegression,
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=True)

# Fit grid hyperparameter search model
gs_log_reg.fit(x_train, y_train);
# Check the best parameters
gs_log_reg.best_params_
```

```python
# Evaluate the model
gs_log_reg.score(x_test, y_test)
```

clearly we can see the normal LogisticRegression and RandomForest tuned with RandomizedSearchCV have the highest accuracy 0.8688524590163934
So we will check other aspects of each to choose a better model

```python
#For LogisticRegression
print("Logistic Regression :")
print("Accuracy =",accuracy_score(y_test,yp_LogisticRegression))
print("Precision =",precision_score(y_test,yp_LogisticRegression))
print("Recall =",recall_score(y_test,yp_LogisticRegression))
print("F1 =",f1_score(y_test,yp_LogisticRegression))
```

```python
#For RandomForest Tuned with RandomizedSearchCV
print("Random Forest (RandomizedSearchCV) :")
yp_rfRSCV=rs_rf.predict(x_test)
print("Accuracy =",accuracy_score(y_test,yp_rfRSCV))
print("Precision =",precision_score(y_test,yp_rfRSCV))
print("Recall =",recall_score(y_test,yp_rfRSCV))
print("F1 =",f1_score(y_test,yp_rfRSCV))
```

Clearly we can see that tuned random forest have a higher recall and similar accuracy as compared to LogisticRegression
So in heart disease classification higher recall means less number of false negative and our aim in such problems is also minimising false negatives
So we choose Hyper-parameter tuned random forest classifier

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(rs_rf,x_test,y_test)
```

```python
print(confusion_matrix(y_test,yp_rfRSCV))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,yp_rfRSCV))
```

```python
from sklearn.model_selection import cross_val_score

c=RandomForestClassifier(n_estimators=210,
                        min_samples_split=4,
                        min_samples_leaf=19,
                        max_depth=3)
```

```python
cv_acc=cross_val_score(c,x,y,cv=5,scoring="accuracy")
```

```python
cv_acc
```

```python
cv_acc=np.mean(cv_acc)
cv_acc
```

```python
print("Accuracy =",cv_acc)
cv_pre=cross_val_score(c,x,y,cv=5,scoring="precision")
print("Precision =",cv_pre.mean())
cv_recall=cross_val_score(c,x,y,cv=5,scoring="recall")
print("Recall =",cv_recall.mean())
cv_f1=cross_val_score(c,x,y,cv=5,scoring="f1")
print("F1 Score =",cv_f1.mean())
```

```python
# Visualizing cross-validated metrics
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                            "Precision": cv_pre.mean(),
                            "Recall": cv_recall.mean(),
                            "F1": cv_f1.mean()},
                          index=[0])
cv_metrics.T.plot.bar(title="Cross-Validated Metrics", legend=False);
```
